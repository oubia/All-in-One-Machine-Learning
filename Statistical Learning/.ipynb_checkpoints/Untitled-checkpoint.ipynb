```python
def vander(X,k):
    n=X.shape[0]
    exp=range(k)
    result=np.zeros((n,k))
    for i in range(k):
        result[:,i]=np.power(X,exp[i])
    return result
```

• Write a function that takes as input the training data D = (X, Y ) and K and returns the MLE
solution (with Gaussian assumption) θMLE ∈ R
K for that problem. Note that the loss function
can be optimized by GD, SGD or Normal Equations.

```python
def SGD(f, grad_f,k,teta0,regularization,X,Y,n_epochs,batch_size,variance_2,alpha,kmax):
    # Initialize w
    tetaSol = teta0 # (K,1)
    # Initialize a vector (loss)
    loss_vec = np.zeros((n_epochs,) ) # (n_epochs,)
    # Extract X and y
    # Save in memory the length of X 
    N = X.shape[0] #
    # Compute the number of batches per epoch
    n_batch_per_epoch = N // batch_size
    # Iterate over the epochs
    x=[]
    for epoch in range(n_epochs):
        idx = np.arange(N) # 1000
        np.random.shuffle(idx)
        # Batch iteration
        for i in range(n_batch_per_epoch): #10 times
            X_temp = X[idx[i * batch_size : (i+1) * batch_size]] # batch_sizex1
            y_temp = Y[idx[i * batch_size : (i+1) * batch_size]] # batch_sizex1
            B = (X_temp, y_temp)
            # Gradient descent update 
            tetaSol = tetaSol - alpha * grad_f(X,Y,k,tetaSol,regularization,variance_2)
            x.append(tetaSol)
    return x
def GDWithoutBackTracking(f, grad_f,k,teta0,regularization,X,Y,variance_2,alpha, kmax):
    tetaSol=teta0
    teta=[teta0]
    f_val=[f(X,Y,k,tetaSol,regularization,variance_2)]
    grad=[grad_f(X,Y,k,tetaSol,regularization,variance_2)]
    err=[np.linalg.norm(grad_f(X,Y,k,tetaSol,regularization,variance_2))]
    i=0
    cond1=True
    cond2=True
    x=[]
    while i < kmax :
    #and (np.linalg.norm(grad_f(xsol),2))>=tolf*np.linalg.norm(grad_f(x0),2) and np.linalg.norm(x[k]-x[k-1],2) >= tolx :
        tetaSol=(tetaSol-alpha*grad_f(X,Y,k,tetaSol,regularization,variance_2))
        x.append(tetaSol)
        f_val.append(f(X,Y,k,tetaSol,regularization,variance_2))
        grad.append(grad_f(X,Y,k,tetaSol,regularization,variance_2))
        err.append(np.linalg.norm(grad_f(X,Y,k,tetaSol,regularization,variance_2))) 
        i+=1
    return x,k,f_val,grad,err
def f(X, Y, k,teta,regularization, variance_2):
    phi=vander(X,k)
    error_vector = phi @ teta - Y
    squared_error = np.linalg.norm(error_vector,2)**2
    return 1 / (2 * variance_2) * squared_error
 
def grad_f(X,Y,k,teta,regularization,variance_2):
    #phi =vander(X,k) # Assuming X is already the matrix of design (e.g., Vandermonde matrix)
    phi=vander(X,k)
    error_vector = phi @ teta - Y
    gradient_vector = (1 / variance_2) * phi.T @ error_vector
    return gradient_vector
```

```python
def avg_abs_error(teta,Test_set):
    X_test,Y_test=Test_set
    k=teta.shape[0]
    N = X_test.shape[0]
    phi=vander(X_test,k)
    Y_pred=phi@teta
    return (np.linalg.norm(Y_pred-Y_test,2)**2)/N
```

```python
def plot_learnt_model_MLE(D,Test_Set,K,method,alpha=None,n_epochs=None,batch_size=None):
    X_train,Y_train=D
    X_test,Y_test=Test_Set
    fig,ax=plt.subplots(int(len(K)/2),2,figsize=(10,12))
    row=0
    col=0
    for i in K:
        ax[row,col].plot(sorted(X_train),sorted(Y_train),"o",markersize = 1,color="black",label="Train Data")
        ax[row,col].plot(sorted(X_test),sorted(Y_test),"o",markersize = 1,color="green",label="Test Data")
        phi=vander(X_train,i)
        teta0=np.zeros(i)
        Y_pred=phi@MLE_solution(D,i,method,alpha,teta0,n_epochs,batch_size)
        ax[row,col].plot(sorted(X_train), sorted(Y_pred),color="violet",label="Predicted Data Train")
        ax[row,col].legend()
        ax[row, col].set_title(f'K = {i}')
        col=(col+1)%2
        if col==0:
            row+=1
    plt.suptitle(f"MLE regression with increasing dimension of parameters K (Real dimension K={k}) using {method}")
    plt.tight_layout()
    plt.show()
```

```python
def f2(X,Y,k,teta,regularization,variance_2):
    phi=vander(X,k)
    return (np.linalg.norm(phi@teta-Y,2)**2 + (regularization*np.linalg.norm(teta,2)**2))/2
def grad2(X,Y,k,teta,regularization,variance_2):
    phi=vander(X,k)
    return phi.T@(phi@teta-Y) + regularization*teta
def normal_equations_minimizer_MAP(X,Y,k,regularization):
    I = np.diag(np.diag(np.ones((k,k))))
    phi=vander(X,k)
    return np.linalg.inv(phi.T @ phi + regularization*I) @ phi.T @ Y
def MAP_solution(D,k,regularization,minimizer_function,alpha=None,teta0=np.zeros(k),n_epochs=None,batch_size=None):
    X,Y=D
    if minimizer_function=="GD":
        x,k,f_val,grad,err=GDWithoutBackTracking(f2, grad2,k,teta0,regularization,X,Y,variance_2,alpha, 1000)
        return x[-1]
    if minimizer_function=="normal equation":
        return normal_equations_minimizer_MAP(X,Y,k,regularization)
    if minimizer_function=="SGD":
        x=SGD(f2, grad2,k,teta0,regularization,X,Y,n_epochs,batch_size,variance_2,alpha, 1000)
        return x[-1]
 #i'll use SGD to find the solution
teta1=MAP_solution(D,10,0.001,"normal equation",1e-4)
teta2=MAP_solution(D,10,100,"normal equation",1e-4)
```

• For K lower, equal and greater than the correct degree of the test polynomial, plot the training
atapoints and the test datapoints with different colors, and visualize (as a continuous line) the
learnt regression model fθMAP (x) with different values of λ. Comment the results.


```python
def plot_learned_model_MAP(D,Test_Set,K,regularizations,method,alpha=None,n_epochs=None,batch_size=None):
    X_train,Y_train=D
    X_test,Y_test=Test_Set
    fig, ax = plt.subplots(len(regularizations), len(K), figsize=(25, 25))
    col=0
    row=0
    for i in K:
        for j in regularizations:
            zipped_Train = sorted(zip(X_train,Y_train), key=lambda x: x[1])
            X_train_ordered,Y_train_ordered= zip(*zipped_Train)
            ax[row,col].plot(X_train_ordered,Y_train_ordered,"o",markersize = 1,color="black",label="Train Data")
            zipped_Test = sorted(zip(X_test,Y_test), key=lambda x: x[1])
            X_test_ordered,Y_test_ordered= zip(*zipped_Test)
            ax[row,col].plot(X_test_ordered,Y_test_ordered,"o",markersize = 1,color="darkblue",label="Test Data")
            phi=vander(X_train,i)
            teta0=np.zeros(i)
            Y_pred=phi@MAP_solution(D,i,j,method,alpha,teta0,n_epochs,batch_size)
            zipped_Train_Pred = sorted(zip(X_train,Y_pred), key=lambda x: x[1])
            X_train_ordered,Y_pred_ordered= zip(*zipped_Train_Pred)
            ax[row,col].set_title(f'Plot with k = {i} and lambda= {j} with {method}')
            ax[row,col].plot(X_train_ordered, Y_pred_ordered,color="green",label="Predicted Data Train")
            ax[row,col].legend()
            col=(col+1)%5
            if col==0:
                row+=1
    plt.tight_layout()
    plt.suptitle(f"Regression with MAP with increasing K and for each K increasing Lambda using {method}")
    plt.show()
```

||θtrue||2
, where θtrue
has been padded with zeros to match the shape of θ. Compute Err(θMLE) and Err(θMAP ) for
increasing values of K and different values of λ.

```python
def err(teta,teta_true):
    new_teta=np.zeros(len(teta))
    teta_padded =np.append(teta_true,np.zeros(len(teta)-len(teta_true)))
    return np.linalg.norm(teta-teta_padded,2)/np.linalg.norm(teta_padded,2)
 
def plot_Error_MLE_MAP(D,K,regularizations,teta_true,method,alpha=None,n_epohcs=None,batch_size=None):
    fig,ax= plt.subplots(len(regularizations),1,figsize=(12,12))
    col=0
    row=0
    cnt=0
    for j in regularizations:
        error1=[]
        error2=[]
        for i in K:
            teta0=np.zeros(i)
            teta1=MLE_solution(D,i,method,alpha,teta0,n_epohcs,batch_size)
            teta2=MAP_solution(D,i,j,method,alpha,teta0,n_epohcs,batch_size)
            error1.append(err(teta1,teta_true))
            error2.append(err(teta2,teta_true))
            ax[cnt].plot(K,error1,label="MLE error")
            ax[cnt].plot(K,error2,label="MAP error")
            ax[cnt].set_title(f'Plot of errors with regularization={j} ')
            ax[cnt].legend()
            cnt+=1
            col=(col+1)%1
            if col==0:
                row+=1
    plt.suptitle(f"Error with increasing value of K and for each K increasing value of penalty term with {method}")
    plt.tight_layout()
    plt.show()
```
